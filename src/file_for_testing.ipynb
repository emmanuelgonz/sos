```python
from dotenv import load_dotenv
import os
load_dotenv('C:\\Users\\dramach6\\OneDrive - Arizona State University\\Documents\\sos\\src\\.env')
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import earthaccess
import xarray as xr
import rioxarray as rxr
```

Resolution

```python
# Loading resolution files from the folder
# Defining paths
path_shp = os.getenv('path_shp')
path_preprocessed = os.getenv('path_preprocessed')
file_name_preprocessed = 'preprocessed_resolution'
raw_path = os.getenv('raw_path')
path_efficiency = os.getenv('path_efficiency')
file_name_efficiency = 'efficiency_resolution'
file_name_efficiency_taskable = 'efficiency_resolution_taskable'

res = xr.open_dataset(path_preprocessed + file_name_preprocessed + '.nc')
#res['Monthly_Resolution_Abs'].isel(time = 2).plot()
eff = xr.open_dataset(path_efficiency + file_name_efficiency + '.nc')
eff_task = xr.open_dataset(path_efficiency + file_name_efficiency_taskable + '.nc')
```

```python
selected_time = 2
plt.subplots(figsize=(15, 5))
plt.subplot(1,3,1)
eff['Monthly_Resolution_Abs'].isel(time = selected_time).plot()
plt.title("Efficiency output - resolution")

plt.subplot(1,3,2)
res['Monthly_Resolution_Abs'].isel(time = selected_time).plot()
plt.title("Datasource output - resolution")

plt.subplot(1,3,3)
eff_task['Monthly_Resolution_Abs'].isel(time = selected_time).plot()
plt.title("Efficiency taskable output - resolution")
plt.tight_layout()

plt.show()
```

Snow Cover

```python
# Loading resolution files from the folder
# Defining paths
path_hdf = os.getenv('path_hdf')
path_nc = os.getenv('path_nc')
path_shp = os.getenv('path_shp')
path_preprocessed = os.getenv('path_preprocessed')
file_name_preprocessed = 'preprocessed_snow_cover'
path_efficiency = os.getenv('path_efficiency')
file_name_efficiency = 'efficiency_snow_cover'

sc = xr.open_dataset(path_preprocessed + file_name_preprocessed+'.nc')
eff_sc = xr.open_dataset(path_efficiency + file_name_efficiency + '.nc')
```
